## JAGS

```r
# libraries 
library(extraDistr)
require(mcsm)
data(Energy)
require(rjags)
require(runjags)
```

Prior :

$$
\theta \sim Beta(\alpha,\beta)
$$

Likelihood:

Probability of duration fora given user: 

$$
P(T=t | \theta) = \theta (1-\theta)^{t-1}
$$

Probability of remaining for a given user:
$$
S(t | \theta) = (1-\theta)^t
$$

Likelihood, assuming independence on the different users :

$$
\prod_{t=1}^{T_{max}}P(T=t | \theta)^{n_t}S(T_{max} | \theta)^{N-\sum_{t=1}^{T_{max}}n_t}
$$

#### Model for JAGS -  precision parameters

```r
survival = c(1000, 869, 743, 653, 593, 551, 517, 491)
died = c(0, 131, 126, 90, 60, 42, 34, 26)

members = c(rep(0,died[2]),rep(1,died[3]),rep(2,died[4]),rep(3,died[5]),rep(4,died[6]),rep(5,died[7]),rep(6,died[8]))
N = length(members)


dataList = list(
    members = members,
    N = N
    )

modelString <-"
    model {
    
    for (i in 1:N){
        members[i] ~ dnegbin(theta,1)     
            }
    theta ~ dbeta(alfa, beta)
    alfa ~ dunif(0.0001,1000)
    beta ~ dunif(0.0001,1000)

    }"
# Write model to a file, to be read by JAGS
modelString2 <-"
model {
    # Likelihood:
    for( i in 1 : N ) {
        members[i] ~ dnegbin( theta , r )
    }
    r<-1
    m <- r*(1-theta)/theta
    # Prior:
    theta ~ dbeta(1.001,1.001)
}"




writeLines(modelString,con="TEMPmodel.txt" )
```

#### Initial values

```r
theta.Init<-rep(theta.Init,N)

```

```r
initsList<-list(theta=theta.Init)
```

```r
initsList
```

#### Parameters we are interested in

```r
parameters<-c("theta")
```

#### Call JAGS

```r
numSavedSteps<-20000
adaptSteps<-500     # Number of steps to adapt the samplers
burnInSteps<-500    # Number of steps to burn-in the chains
nChains<-4
thinSteps<-1
nIter<-ceiling( ( numSavedSteps * thinSteps ) / nChains )
# 
jagsModel<-jags.model("TEMPmodel.txt" ,data=dataList,inits=initsList,n.chains=nChains,n.adapt=adaptSteps)
```

#### Burn in

```r
update(jagsModel, n.iter=burnInSteps)
```

```r
codaSamples<-coda.samples(jagsModel ,variable.names=parameters,n.iter=nIter,thin=thinSteps)
```

### Badoink VR data

Let's explain the model :

In order to estimate $\theta$ (probability of failure, i.e. probability of cancel) by means of MCMC, 

```r
dataVR = readRDS("BaDoinkVR.rds")
```

```r
VR_died = dataVR$Diff
VR_survival = dataVR$Count
```

```r
members = c(rep(0,died[2]),rep(1,died[3]),rep(2,died[4]),rep(3,died[5]),rep(4,died[6]),rep(5,died[7]),rep(6,died[8]))
N = length(members)
T_max= 12

VR_died_by_users = 0
init = 1
for (i in 1:12){
    window =  VR_died[i+1]
    VR_died_by_users = c(VR_died_by_users, rep(i-1,window))
    init = init + window
}
```

```r
# frequentist probability estimation
VR_denom = VR_survival[1]
probs = rep(0,length(VR_died)-1)
for (i in 1:(length(VR_died)-1)){
    VR_denom = VR_denom-VR_died[i]
    probs[i]=VR_died[i+1]/VR_denom

}

theta.Init = mean(probs)
theta.Init
```

```r
N = length(VR_died_by_users)

dataList = list(
    VR_died_by_users = VR_died_by_users,
    N = N
    )

modelString <-"
    model {
    
    for (i in 1:N){
        VR_died_by_users[i] ~ dnegbin(theta,1)     
            }
    theta ~ dbeta(alfa, beta)
    alfa ~ dunif(0.0001,1000)
    beta ~ dunif(0.0001,1000)

    }"
```

```r
writeLines(modelString,con="TEMPmodel.txt" )
```

```r
parameters<-c("theta")
```

```r
numSavedSteps<-50000
adaptSteps<-500     # Number of steps to adapt the samplers
burnInSteps<-500    # Number of steps to burn-in the chains
nChains<-2
thinSteps<-1
nIter<-ceiling( ( numSavedSteps * thinSteps ) / nChains )
# 
jagsModel<-jags.model("TEMPmodel.txt" ,data=dataList,inits=initsList,n.chains=nChains,n.adapt=adaptSteps)
```

```r
update(jagsModel, n.iter=burnInSteps)
```

```r
codaSamples<-coda.samples(jagsModel ,variable.names=parameters,n.iter=nIter,thin=thinSteps)
```
